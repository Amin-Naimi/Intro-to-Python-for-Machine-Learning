### . Data Visualization Library

```python
import seaborn as sns
import matplotlib.pyplot as matplt
import pandas as pd
import numpy as np
```

### Note : Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips') # Pandas data frame
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()#Default theme
# visualize the data
sns.relplot(data = tips, x ='total_bill',y='tip',col="sex",hue='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

### Scatter Plot:

```python
sns.scatterplot(x ='sepal_length', y = 'petal_length', hue="species", data=iris )
```

```python
# loading the titanic data set
titanic = sns.load_dataset("titanic")
titanic.head()
```

### Count Plot

```python
sns.countplot(x = 'class', data = titanic)
```

```python
sns.countplot(x="survived",data=titanic)
```

### Bar Chart

```python
sns.barplot(x="sex",y='survived',hue="class",data=titanic)
```

### house price dataset

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
boston_house = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
boston_house.head()
house = boston_house
house['PRICE'] = np.random.randint(10,100,1012)
house.head()
```

### Distribution Plot

```python
sns.distplot(house['PRICE'])
```

### Correlation :
1. Positive Correlation ( v1 et v2) v1 +++ => v2 +++
2. Negative Correlation (v1 et v2) v1  +++ => v2 ---

```python
#Heat Map
# construction a heat map
correlation = house.corr()
matplt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8},cmap="Blues")
```
