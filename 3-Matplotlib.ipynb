# Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python

```python
import matplotlib.pyplot as matplt
# importing numpy to get data for our plots
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z= np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

### Ploting the data

```python
# sin wave
matplt.plot(x,y)
matplt.show
```

```python
# cosine wave
matplt.plot(x,z)
matplt.show()
```

```python
# adding title x-axis and y-axis labels
matplt.plot(x,y)
matplt.xlabel("angle")
matplt.ylabel("sine value")
matplt.title("sine wave")
matplt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
matplt.plot(x,y)
matplt.show()
```

```python
matplt.plot(x,y, "r+")
matplt.show()
```

```python
matplt.plot(x,y,'g.')
matplt.show()
```

```python
x = np.linspace(-5,5,50)
matplt.plot(x,np.sin(x), 'g-')
matplt.plot(x,np.cos(x),"r--")
```

### Bar Plot

```python
figure = matplt.figure()
ax = figure.add_axes([0,0,1,1])
# 0 -> corrdinates (origine)
# 0 -> corrdinates (origine)
# 1 -> height of the rectangle
# 1 -> widht of the rectangle
languages = ["Arabic","French","Spanish","Latin","English"]
people = [1000,50,20,80,500]
ax.bar(languages,people)
matplt.xlabel("Languages")
matplt.ylabel("People")
figure.show()
```

### Pie Chart

```python
figure = matplt.figure()
ax = figure.add_axes([0,0,1,1])
languages = ["English", "Arabic","French","Lating","Spanish"]
people = [20,500,10,30,15]
ax.pie(people,labels=languages, autopct='1%.1f%%')
figure.show()
```

### Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
figure_2 = matplt.figure()
ax = figure_2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
matplt.show()
```

### 3D Scatter Plot

```python
figure = matplt.figure()
ax =  matplt.axes(projection="3d")
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
matplt.show()
```
